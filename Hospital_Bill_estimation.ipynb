```python
def scale(data_frame,stri):
    x = list(data_frame[stri])
    mn = min(x)
    mx = max(x)
    l=[]
    for i in x:
        l.append((i-mn)/(mx-mn))
    return mx,l
```

```python
max_age,l_age = scale(data_frame2,'Age')
max_los,l_los = scale(data_frame2,'LOS')
max_bill,l_bill = scale(data_frame2,'Bill Amt')

data_frame2.insert(1,"bill",l_bill)
data_frame2.insert(1,"los",l_los)
data_frame2.insert(1,"age",l_age)
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(data_frame, test_size=0.05)
```

```python
y_train = np.array(train['bill'])
y_test = np.array(test['bill'])
```

```python
x_train = np.array(train)
x_train = np.delete(x_train,2,1)
x_test = np.array(test)
x_test = np.delete(x_test,2,1)
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
model = MLPRegressor(activation='tanh', alpha=0.06, batch_size='auto',
                      beta_1=0.9, beta_2=0.999, early_stopping=False,
                      epsilon=1e-08, hidden_layer_sizes=(200,),
                      learning_rate='constant', learning_rate_init=0.01,
                      max_iter=600, momentum=0.6, n_iter_no_change=10,
                      nesterovs_momentum=True, power_t=0.5, random_state=1,
                      shuffle=True, solver='adam', tol=0.0001,
                      validation_fraction=0.1, verbose=False, warm_start=False)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
